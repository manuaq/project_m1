```python
text_rows=[]

for i in soup.select("td"):
    text_rows.append(i.text)

print(text_rows)
```

```python
clean_rows=[]

for i in iter(text_rows):
    clean_rows = [i.replace('\n','') for i in text_rows]

print(clean_rows)
```

```python
without_empty_strings = [string for string in clean_rows if string != ""]
```

```python
iter_strings = iter(without_empty_strings)
```

```python
without_empty_strings
```

```python
country_code_name_pairs = [h + next(iter_strings, '') for h in iter_strings]

country_code_name_pairs
```

```python
len(country_code_name_pairs)
```

```python
country_code_name_pairs = [i for i in country_code_name_pairs if i  != ' \xa0 \xa0']

print(country_code_name_pairs)
```

```python
country_code_name_pairs = {x.replace("[1]", "") for x in country_code_name_pairs}
country_code_name_pairs = {x.replace("[2]", "") for x in country_code_name_pairs}
country_code_name_pairs = {x.replace("*", "") for x in country_code_name_pairs}

country_code_name_pairs
```

```python
country_df = pd.DataFrame(country_code_name_pairs)

country_df
```

```python
country_df.columns =['country_name_and_acronym']

country_df
```

```python
country_df['dirty_country_acronym'] = country_df['country_name_and_acronym'].str.extract('(\(.+\))', expand=True)

country_df
```

```python
country_df['dirty_country_name'] = country_df['country_name_and_acronym'].str.extract('(^.+?(?=\())', expand=True)

country_df.tail(10)
```

```python
country_df['country_name'] = country_df['dirty_country_name'].str.lstrip().str.rstrip()

country_df['country_name']
```

```python
country_df['country_name'].unique()
```

```python
country_df['country_acronym'] = country_df['dirty_country_acronym'].str.extract('([A-Z]+)', expand=True)

country_df
```

```python
country_df = country_df[['country_name','country_acronym']]

country_df
```
